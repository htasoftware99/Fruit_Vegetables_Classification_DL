```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense,Conv2D,Flatten,MaxPooling2D,Input,Dropout,BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.layers import LeakyReLU
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
data_train_path ="data/train"
data_test_path="data/test"
data_val_path = "data/validation"
```

```python
img_width = 180
img_height = 180
```

```python
data_train = tf.keras.utils.image_dataset_from_directory(
data_train_path,
    shuffle= True,
    image_size= (img_width, img_height),
    batch_size = 32,
    validation_split=False
)
```

```python
import os

# Dizin yolunu kontrol et
data_train_path = 'data/test'  # Örnek bir yol, kendi yolunuza göre güncelleyin
if not os.path.exists(data_train_path):
    print(f"Dizin bulunamadı: {data_train_path}")
else:
    print(f"Dizin mevcut: {data_train_path}")
```

```python
data_cat = data_train.class_names
```

```python
data_val = tf.keras.utils.image_dataset_from_directory(
    data_val_path,
    image_size=(img_width,img_height),
    batch_size=32,
    shuffle=False,
    validation_split=False
)
```

```python
data_test = tf.keras.utils.image_dataset_from_directory(
data_test_path,
    image_size=(img_width,img_height),
    batch_size=32,
    shuffle=False,
    validation_split=False
)
```

```python
plt.figure(figsize=(10,10))
for image, labels in data_train.take(1):
    for i in range(9):
        plt.subplot(3,3,i+1)
        plt.imshow(image[i].numpy().astype('uint8'))
        plt.title(data_cat[labels[i]])
        plt.axis('off')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
```

```python
from tensorflow.keras.layers import Rescaling

# Define the rescaling layer
rescaling = Rescaling(1./255)

# Define the Sequential model
model = Sequential([
    rescaling,  # Rescaling layer
    Conv2D(16, 3, padding='same', activation='relu'),
    MaxPooling2D(),
    Conv2D(32, 3, padding='same', activation='relu'),
    MaxPooling2D(),
    Conv2D(64, 3, padding='same', activation='relu'),
    MaxPooling2D(),
    Flatten(),
    Dropout(0.2),
    Dense(128),
    Dense(len(data_cat)),
])
```

```python
model.compile(optimizer='adam', loss =tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
```

```python
history = model.fit(data_train, validation_data=data_val,epochs=25,batch_size=32,verbose=1 )
```

```python
import seaborn as sns
sns.set(style="darkgrid")
epochs=25
plt.plot(np.arange(0,epochs), history.history["loss"], label="train_loss")
plt.plot(np.arange(0,epochs), history.history["val_loss"], label="val_loss")
plt.plot(np.arange(0,epochs), history.history["accuracy"], label="train_accruracy")
plt.plot(np.arange(0,epochs), history.history["val_accuracy"], label="val_accuracy")
plt.title("model grafiği")
plt.xlabel("epochs")
plt.ylabel("loss & acc")
plt.legend()
plt.show()
```

```python
model.save("Fruit_Vegetable_Model.keras")
```

```python
from tensorflow.keras.models import load_model

# Modeli yükle
model = load_model("Fruit_Vegetable_Model.keras")
```

```python
from tensorflow.keras.preprocessing import image
```

```python
def prepare_image(image_path, target_size=(180, 180)):
    # Resmi yükle
    img = image.load_img(image_path, target_size=target_size)
    
    # Resmi numpy dizisine dönüştür
    img_array = image.img_to_array(img)
    
    # Ek boyut ekle (batch dimension)
    img_array = np.expand_dims(img_array, axis=0)
    
    # Normalizasyon
    img_array = img_array / 255.0
    
    return img_array
```

```python
def predict_image(image_path):
    img_array = prepare_image(image_path)
    predictions = model.predict(img_array)
    return predictions

# Test resmi
image_path = '1.jpg'  # Test resminizin yolu
predictions = predict_image(image_path)
```

```python
def get_prediction_result(predictions, class_names):
    # Tahminlerin olasılıklarını alın
    probabilities = tf.nn.softmax(predictions[0])
    
    # En yüksek olasılığa sahip sınıfı bulun
    predicted_class = np.argmax(probabilities)
    
    # Olasılık yüzdesini hesaplayın
    confidence = probabilities[predicted_class] * 100
    
    return class_names[predicted_class], confidence

# Sınıf isimlerini alın
class_names = data_train.class_names

# Tahmin sonucunu alın
predicted_label, confidence = get_prediction_result(predictions, class_names)
print(f"Tahmin Edilen Etiket: {predicted_label}")
print(f"Güven Skoru: {confidence:.2f}%")
```
